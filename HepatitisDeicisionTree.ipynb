```python
question_mark_columns = ["STEROID" , "FATIGUE" , "MALAISE" , "ANOREXIA" , "LIVER BIG" , "LIVER FIRM" , "SPLEEN PALPABLE" , "SPIDERS" ,
                         "ASCITES" , "VARICES" , "BILIRUBIN" , "ALK PHOSPHATE" , "SGOT" , "ALBUMIN" , "PROTIME"]
```

```python
def MissingFiller(Df , question_mark_columns):
    for i in question_mark_columns : 
        Df[i].replace('?', '0', inplace=True)
        Df[i] = Df[i].astype(np.float32)
        Df[i].replace(0, Df[i].mean(), inplace=True)
        
    return Df
```

```python
Df = MissingFiller(Df , question_mark_columns)
```

```python
Df.head()
```

Building Feature set

```python
X = Df.iloc[:,1:].values.astype(np.float32)
Y = Df.iloc[:,0].values.astype(np.int32)
```

```python
print(X.shape)
print(Y.shape)
```

Train Test Split 

```python
from sklearn.model_selection import train_test_split 

Xtrain , Xtest , Ytrain , Ytest = train_test_split(X , Y , test_size = 0.25 , random_state = 2)

print(Xtrain.shape)
print(Xtest.shape)
print(Ytrain.shape)
print(Ytest.shape)
```

Standardization

```python
from sklearn.preprocessing import StandardScaler 

Scale = StandardScaler()

Xtrain = Scale.fit_transform(Xtrain)
Xtest = Scale.transform(Xtest)
```

```python
print(Xtrain.shape)
print(Xtest.shape)
```

Building model

```python
def Score(model , X , Y):
    Ypred = model.predict(X)
    return (Ypred == Y).mean()
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(Xtrain , Ytrain)

print(" Accuracy : " , Score(model , Xtest , Ytest))
```

Finding Best Value for max-depth

model improvement 

```python
# Finding max - depth between 1 and 100 

for i in range(1,101):
    model = DecisionTreeClassifier(max_depth = i)
    model.fit(Xtrain , Ytrain)
    print("Training Accuracy : " , Score(model , Xtrain , Ytrain))
    print("Test Accuracy at Max-Depth : " , i , " Is : " , Score(model , Xtest , Ytest))
    print("\n")
```

Best Testing Accuracy is 82.05%
